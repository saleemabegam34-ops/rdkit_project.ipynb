```python
import pandas as pd
```

```python
from rdkit import Chem
```

```python
from rdkit.Chem import Draw
```

```python
data={'compounds': ['Aspirin', 'Caffeine'], 'SMILES':['CC(=O)OC1=CC=CC=C1C(=O)O', 'CN1C=NC2=C1C(=O)N(C(=O)N2C)C']}
df=pd.DataFrame(data)
```

```python
print(df)
```

```python
def smiles_to_mol(smiles):
    mol=Chem.MolFromSmiles(smiles)
    return mol
```

```python
df['Molecule'] = df['SMILES'].apply(smiles_to_mol)
```

```python
print(df.head())
```

```python
molecules=df['Molecule'].tolist()
```

```python
img = Draw.MolsToGridImage(molecules, legends=df['compounds'].astype(str).tolist(), molsPerRow=3,subImgSize=(200, 200))

```
